```python
import ipywidgets as widgets
from IPython.display import display
from traitlets import traitlets

# Define the calculator functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

# Create input widgets
num1_input = widgets.FloatText(description="First number:", style={'description_width': 'initial'})
operator_input = widgets.Dropdown(options=["+", "-", "*", "/"], description="Operator:", style={'description_width': 'initial'})
num2_input = widgets.FloatText(description="Second number:", style={'description_width': 'initial'})
calculate_button = widgets.Button(description="Calculate", button_style='success')

# Create output widget
result_output = widgets.Output()

# Define calculation function
def calculate_button_clicked(b):
    with result_output:
        # Retrieve user input
        num1 = num1_input.value
        operator = operator_input.value
        num2 = num2_input.value

        # Perform calculation based on the operator
        if operator == "+":
            result = add(num1, num2)
        elif operator == "-":
            result = subtract(num1, num2)
        elif operator == "*":
            result = multiply(num1, num2)
        elif operator == "/":
            if num2 != 0:
                result = divide(num1, num2)
            else:
                result = "Cannot divide by zero"
        else:
            result = "Invalid operator"

        # Display the result
        print("Result:", result)

# Register click event for the calculate button
calculate_button.on_click(calculate_button_clicked)

# Display the widgets
display(widgets.VBox([num1_input, operator_input, num2_input, calculate_button, result_output]))
```
